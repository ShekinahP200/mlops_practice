```python
import matplotlib.pyplot as plt

total_sales_by_product = df.groupby('Product Category')['Total Revenue'].sum()
total_sales_by_product.plot(kind='bar', color='teal')
plt.title('Total Sales by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Revenue')
plt.show()
```

```python
sales_trend = df.groupby('Date')['Total Revenue'].sum().reset_index()
sales_trend
```

```python
plt.plot(sales_trend['Date'],sales_trend['Total Revenue'])
```
